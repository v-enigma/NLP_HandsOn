```python
from PyPDF2 import PdfReader
reader = PdfReader("C:\\Users\\venu9\\OneDrive\\Desktop\\Research Papers\\1301.3781v3.pdf")
text =[]
words =[]
wordDict = dict()
for page in reader.pages:
    text.append(page.extract_text())
for item in text:
    temp = item.split()
    for word in temp:
        words.append(word)
        if word in wordDict:
            wordDict[word] = wordDict[word]+1
        else:
            wordDict[word] = 1
        
#print(words)
print(wordDict)
print(len(words))    
```
